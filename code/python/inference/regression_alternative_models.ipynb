```python
import numpy as np
import pandas as pd
import evaluate_heatmaps as e
import regression_analysis as r
```

# Uniform Baseline

```python
uniform_pred_tr = np.ones((501, 601))
uniform_pred_tr /= np.sum(uniform_pred_tr)
uniform_pred_tr = uniform_pred_tr.ravel()

uniform_pred = np.tile(uniform_pred_tr, 150)
```

```python
print("Vision")
uniform_vision_emd = e.compute_emd_all_trials(uniform_pred, "vision")
print()
print("Audio")
uniform_audio_emd = e.compute_emd_all_trials(uniform_pred, "audio")
print()
print("Occluded")
uniform_occluded_emd = e.compute_emd_all_trials(uniform_pred, "occluded")
```

```python
print("Vision Mean EMD:", np.mean(uniform_vision_emd))
print("Audio Mean EMD:", np.mean(uniform_audio_emd))
print("Occluded Mean EMD:", np.mean(uniform_occluded_emd))
```

```python
path = "model_performance/emd/"

df_uniform_baseline_vision = pd.DataFrame({
    "trial": e.world_nums,
    "distance": uniform_vision_emd
})

df_uniform_baseline_audio = pd.DataFrame({
    "trial": e.world_nums,
    "distance": uniform_audio_emd
})

df_uniform_baseline_occluded = pd.DataFrame({
    "trial": e.world_nums,
    "distance": uniform_occluded_emd
})

df_uniform_baseline_vision.to_csv(path + "baseline_uniform_vision_emd.csv", index=False)
df_uniform_baseline_audio.to_csv(path + "baseline_uniform_audio_emd.csv", index=False)
df_uniform_baseline_occluded.to_csv(path + "baseline_uniform_occluded_emd.csv", index=False)
```

# Visual Features Heuristic

```python
print("Vision")
_, vf_pred_vision, _, _ = r.compute_regression(None, "visual_features", "vision", cut=300)
vf_vision_emd = e.compute_emd_all_trials(vf_pred_vision, "vision", cut=300)
print()

print("Audio")
_, vf_pred_audio, _, _ = r.compute_regression(None, "visual_features", "audio", cut=300)
vf_audio_emd = e.compute_emd_all_trials(vf_pred_audio, "audio", cut=300)
print()

print("Occluded")
_, vf_pred_occluded, _, _ = r.compute_regression(None, "visual_features", "occluded", cut=300)
vf_occluded_emd = e.compute_emd_all_trials(vf_pred_occluded, "occluded", cut=300)
```

```python
path = "model_performance/emd/"

df_vf_vision = pd.DataFrame({
    "trial": e.world_nums,
    "distance": vf_vision_emd
})

df_vf_audio = pd.DataFrame({
    "trial": e.world_nums,
    "distance": vf_audio_emd
})

df_vf_occluded = pd.DataFrame({
    "trial": e.world_nums,
    "distance": vf_occluded_emd
})

df_vf_vision.to_csv(path + "visual_features_vision_emd.csv", index=False)
df_vf_audio.to_csv(path + "visual_features_audio_emd.csv", index=False)
df_vf_occluded.to_csv(path + "visual_features_occluded_emd.csv", index=False)
```
